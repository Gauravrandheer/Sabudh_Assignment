**MatplotlibPractice**

```python
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
import numpy as np
x = np.linspace(0,5,11)
y = x**2
```

```python
x
```

```python
y
```

```python
plt.plot(x,y)
plt.xlabel("X-Label")
plt.ylabel("Y-Label")
plt.title("Experiment")
```

```python
plt.subplot(2,1,1)#row col diagram number
plt.plot(x,y,"r")
plt.subplot(2,1,2)
plt.plot(x,y,"b")
```

```python
# Object oriented method
fig = plt.figure()

axes1 = fig.add_axes([0.1,0.3,0.8,0.8])
axes2 = fig.add_axes([0.5,0.4,0.4,0.3])# left, top bottom, width ,height

axes1.plot(x,y)
axes1.set_title('Larger plot')
axes2.plot(y,x)
axes2.set_title("Small plot")
```

```python
fig,axes = plt.subplots(nrows=1,ncols=2)

axes[0].plot(x,y)
axes[1].plot(y,x)
```

```python
fig = plt.figure(figsize=(8,2))

ax = fig.add_axes([0,0,1,1])
ax.plot(x,y)
```

```python
fig,axes = plt.subplots(nrows=2,ncols=1)
axes[0].plot(x,y)
axes[1].plot(y,x)
plt.tight_layout()
```

```python
fig.savefig("my_picture.jpg",dpi =200)#dots per inch
```

```python
fig = plt.figure()
ax = fig.add_axes([0,0,1,1])
ax.plot(x,y)
ax.plot(y,x)
```

```python
fig = plt.figure()
ax = fig.add_axes([0,0,1,1])
ax.plot(x,x**2,label ="X Square")
ax.plot(x,x**3,label ="X cube")
ax.legend()
```

```python
fig = plt.figure()
ax = fig.add_axes([0,0,1,1])
ax.plot(x,x**2,color="orange",linewidth = 1,alpha=.5,linestyle="-",marker ="o",markersize ="20",markerfacecolor="blue")#RGB Hex Code
ax.set_xlim([0.2,.6])
```
