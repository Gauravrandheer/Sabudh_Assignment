# Pandas Built-in Data Visualization Practice

```python
import numpy as np
import pandas as pd
%matplotlib inline
import seaborn as sns
df = sns.load_dataset("tips")
```

```python
df["total_bill"].hist()
```

```python
df["total_bill"].plot(kind="hist")
```

```python
df["total_bill"].plot.area()
```

```python
df["total_bill"].plot.bar(stacked=True)
```

```python
df["total_bill"].plot.line(x=df.index,y="total_bill",lw=2)
```

```python
df.plot.scatter(x="total_bill",y="tip")
df.head()
```

```python
df.plot.scatter(x="total_bill",y="tip",s=df["size"]*100)
```

```python
df.plot.box()
```

```python
df.plot.hexbin(x="tip",y="total_bill",gridsize=25)
```

```python
df["total_bill"].plot.kde()
```

```python
df["total_bill"].plot.density()
```

```python
 df.plot.density()
```
