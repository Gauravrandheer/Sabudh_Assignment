**Linear regression Using Class **
<h8>y = mx +c +e<h8>
<h8>m= Slope<h8>
<h8>c= Intercept<h8>
<h8>x = random_data_points<h8>
<h8>e = error (normal distrubtion around 0)<h8>

**Using Class Regression**

```python
import random
import numpy as np
import matplotlib.pyplot as plt

class Regression():
    
    def __init__(self,x,e): 
        self.x = x
        self.e = e
    
    def cal_y(self,m,c):
        self.m = m
        self.c = c
        print(m)
        print(c)
        y= m*(self.x) + c + self.e
        print(y)
        plt.scatter(self.x,y)
        plt.show()
        return y

    def cal_y1(self,m,c):
        self.m = m
        self.c = c
        print(m)
        print(c)
        y1= m*(self.x) + c
        print(y)
        plt.scatter(self.x,y)
        plt.show()
        return y1
        
    def solve(self,y):
        self.y = y
        m=2
        c=10
        x = self.x
        for i in range(1,100):
            dcostdm =(y- m*x -c)*x
            dcostdm =-dcostdm/100
            dcostdm =np.sum(dcostdm)
            dcostdc =(y- 2*x -10)
            dcostdc = -dcostdc/100
            dcostdc =np.sum(dcostdc)
            m = m -(0.01)*(dcostdm)
            c = c - (0.01)*dcostdc
        print("Slope:"+str(m))
        print("Intercept:"+str(c))
        return m,c
    
    def plot_graph(self,y,y1):
        self.y = y
        self.y1 = y1
        x = self.x 
        plt.scatter(x,y)
        plt.plot(x,y1)
        plt.show()
        
    def plot_hist(self,y,y1):
        self.y = y
        self.y1= y1
        plt.hist(y1-y)
        
    
x = np.random.normal(loc=0.0,scale=1.0,size=100)*10
m = random.randint(1,10)
c = random.randint(1,10)
e = np.random.normal(loc=0.0,scale=1.0,size=100)*10
```

```python
k = Regression(x,e)

```

```python
y = k.cal_y(m,c)
```

```python
m,c = k.solve(y)
```

```python
y1 = k.cal_y1(m,c)
```

```python
k.plot_graph(y,y1)
```
