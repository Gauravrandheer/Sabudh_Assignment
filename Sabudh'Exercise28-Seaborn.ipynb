```python
import seaborn as sns
%matplotlib inline
```

```python
tips = sns.load_dataset("tips")
```

```python
tips.head()
```

```python
sns.distplot(tips["total_bill"],kde =False,bins = 30)
```

```python
sns.jointplot(x="total_bill",y="tip",data = tips)#kind = reg 
```

```python
sns.pairplot(tips,hue="sex")
```

```python
sns.rugplot(tips["total_bill"])
```

```python
sns.distplot(tips["total_bill"])#kde kernal density estimation
```

#  Categorial data

```python
import numpy as np
sns.barplot(x="sex",y="total_bill",data = tips,estimator=np.std)
```

```python
sns.countplot(x ="sex",data=tips)
```

```python
sns.boxplot(x="day",y="total_bill",data=tips,hue='smoker')
```

```python
sns.violinplot(x="day",y="total_bill",data=tips,hue='smoker',split =True)
```

```python
sns.stripplot(x="day",y="total_bill",data=tips ,jitter=True,hue='smoker')
```

```python
sns.swarmplot(x="day",y="total_bill",data=tips,hue='smoker')
```

```python
sns.boxplot(x="day",y="total_bill",data=tips,hue='smoker')
sns.violinplot(x="day",y="total_bill",data=tips,hue='smoker',color ="black")
```

```python
sns.factorplot(x="day",y="total_bill",data=tips,hue='smoker')
```
